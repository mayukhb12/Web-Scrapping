```python
contents= d.find_all('div',class_='quote')

Contents=[]

for content in contents:
    Q=content.find('span', class_='text').text
    Author=content.find('small', class_='author').text
    Tags=content.find('meta')
    tag=Tags.attrs['content']
    Contents.append([Q,Author,tag])
    
print(Contents)    
```

```python
Contents=[]

for i in range(1,10):
    url=f"http://quotes.toscrape.com/page/{i}/"
    response= requests.get(url)
    response=response.content
    soup= BeautifulSoup(response,'html.parser')
    d=soup.find(class_='container')
    contents= d.find_all('div',class_='quote')
    
    for content in contents:
        Q=content.find('span', class_='text').text
        Author=content.find('small', class_='author').text
        Tags=content.find('meta')
        tag=Tags.attrs['content']
        Contents.append([Q,Author,tag])
```

```python
df=pd.DataFrame(Contents, columns=['Quote','Author','Tag'])
```

```python
df.to_csv('Quotes_to_scrape.csv')
```
